# LOG LIKELIHOOD COST FUNCTION (USED IN MICROGRAD)

Loss Function = $$\sum_{\forall x,y}{-log(NN(x)[y_{idx}])}$$

Log function visualization when prediction is very close to 0 for the right class (BAD PREDICTION) - Returns high loss

```python
# plot negative log likelihood function

import numpy as np

import matplotlib.pyplot as plt


x = np.linspace(0, 0.0001**-10, 1000)

nll = np.array([ -np.log(xi) for xi in x])

plt.plot(x, nll)
```

Log function visualization when prediction is very close to 1 for the right class (GOOD PREDICTION) - Returns low loss

```python
# plot negative log likelihood function

import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0.5, 1, 1000)

nll = np.array([ -np.log(xi) for xi in x])

plt.plot(x, nll)
```

# OPTIMIZE COST FUNCTION WITH 1 PARAMETER CHANGE IN NN EXAMPLE

- ONE INPUT = array with length 3, OUTPUT = scalar number
- LAYER 1 



```python
input = np.array([
    [1, 1, 1],
    [2, 2, 2],
    [3, 3, 3],
])
input
```

```python
output = np.array([1,2,3])
output
```

```python
weights = np.array([1, 1, 1])
bias = np.array([0, 0, 0])

result = input.dot(weights) + bias

result
```

```python
cost = np.sum((result - output)**2)
cost 
```

How w1 affects cost?

cost = $$[(w_1*x_1 ...)-output]^2 +[(w_1*x_2 ...)-output]^2+[(w_1*x_3 ...)-output]^2$$

It is clear it affecs it with every sample. 
As 
- cost is calculated as sum of cost for each sample 
- according to the sum rule for derivation the derivative for sum of functions is the sum of their derivates 

Thus we can just sum up the $w_1$ gradient for each sample separately and update it based on each training sample. 

This does not change as we add more layers or neurons to the neural network.
